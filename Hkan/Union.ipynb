```python
import pandas as pd
dfTB = pd.read_csv('ingesta_D2.csv')
```

```python
nulos_por_col = dfTB.isnull().sum()
totales = dfTB.shape[0]
pct_por_col = (nulos_por_col / totales) * 100

print("\nPorcentaje de nulos por columna (%):\n", pct_por_col.round(2))
```

```python
cols_to_drop = pct_por_col[pct_por_col > 25].index.tolist()

dfTB.drop(columns=cols_to_drop, inplace=True)

print(f"Se eliminaron {len(cols_to_drop)} columnas: {cols_to_drop}")
print("Columnas restantes:", dfTB.columns.tolist())
```

```python
# eliminar todas las columnas excepto SEQN y BMXWT
keep = ['SEQN', 'DR2TKCAL', 'DR2TPROT', 'DR2TCARB', 'DR2TSUGR', 'DR2TFIBE', 'DR2TTFAT', 'DR2TSFAT', 'DR2TCHOL', 'DR2TRET']
to_drop_now = [c for c in dfTB.columns if c not in keep]

dfTB.drop(columns=to_drop_now, inplace=True)

# actualizar lista global de columnas eliminadas (si existe)
try:
    cols_to_drop.extend(to_drop_now)
except NameError:
    cols_to_drop = to_drop_now.copy()

print(f"Se eliminaron {len(to_drop_now)} columnas ahora: {to_drop_now}")
print("Columnas restantes:", dfTB.columns.tolist())
```

```python
output_file = 'Entrevista_ingesta_D2.csv'
dfTB.to_csv(output_file, index=False)

print(f"CSV actualizado guardado en: {output_file}")
print(f"Filas: {dfTB.shape[0]}, Columnas: {dfTB.shape[1]}")
print(f"Columnas eliminadas ({len(cols_to_drop)}): {cols_to_drop}")
```
